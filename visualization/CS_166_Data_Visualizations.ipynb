```python
import json
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("honeypot_data.csv")
sns.countplot(data=df, y="protocol", hue="protocol")
plt.title("What Are The Most Common Protocols?")
```

```python
temp = df
temp=temp[temp["protocol"] != "pcap"]
temp=temp[temp["protocol"] != "TCP"]
temp=temp[temp["protocol"] != "smbd"]

sns.countplot(data=temp, y="protocol", hue="protocol")
plt.title("What Are The Most Common Protocols Without PCAP, TCP, SMBD?")
```

```python
groups = df.groupby("destination_port").count()
groups = groups.sort_values("protocol", ascending=False)
groups.head(20)

des_help = groups.head(20)
des_help["Destination Port"] = "Destination Port: " + des_help.index.astype(str)
des_help

sns.barplot(data=des_help, x="protocol", y="Destination Port", hue="Destination Port")
plt.title("What Destination Ports Are Most Commonly Associated With Attacks?")
```

```python
source_groups = df.groupby("source_port").count()
source_groups = source_groups.sort_values("protocol", ascending=False)
source_groups.head(20)

source_help = source_groups.head(20)
source_help["Source Port"] = "Source Port: " + source_help.index.astype(str)
source_help

sns.barplot(data=source_help, x="protocol", y="Source Port", hue="Source Port")
plt.title("What Source Ports Are Most Commonly Associated With Attacks?")
```
